```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Automobile_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.replace({'?':np.nan})
```

```python
fig, ax=plt.subplots(figsize=(12, 8))
ax=sns.heatmap(df.isna())
plt.show()
```

```python
df['price']=df['price'].apply(pd.to_numeric)
df['peak-rpm']=df['peak-rpm'].apply(pd.to_numeric)
df['horsepower']=df['horsepower'].apply(pd.to_numeric)
df['stroke']=df['stroke'].apply(pd.to_numeric)
df['normalized-losses']=df['normalized-losses'].apply(pd.to_numeric)
df['bore']=df['bore'].apply(pd.to_numeric)
```

```python
name_num={'one':1, 'two':2, 'three':3, 'four':4, 'five': 5, 'six':6, 'seven':7, 'eight': 8,'twelve':12}
```

```python
df['num-of-doors']=df['num-of-doors'].replace(name_num)
df['num-of-cylinders']=df['num-of-cylinders'].replace(name_num)
```

```python
df['stroke'].fillna(df['stroke'].median(), inplace=True)
df['peak-rpm'].fillna(df['peak-rpm'].median(), inplace=True)
```

```python
df['num-of-doors']=pd.to_numeric(df['num-of-doors'])
df['num-of-cylinders']=pd.to_numeric(df['num-of-cylinders'])
```

```python
df['num-of-doors'].fillna(df['num-of-doors'].mean(), inplace=True)
df['horsepower'].fillna(df['horsepower'].mean(), inplace=True)
df['bore'].fillna(df['bore'].median(), inplace=True)
```

```python
cat_cols=[]
for i in df.columns:    
    if (df[i].dtype=='object'):
        cat_cols.append(i)
print(cat_cols) 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in cat_cols:
    le=LabelEncoder()
    le.fit(df[i])
    df[i]=pd.Series(le.transform(df[i]))
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X=df.drop(columns=['normalized-losses', 'price'], axis=1)
```

```python
y=df['price']
```

```python
scaler=StandardScaler()
X1=pd.DataFrame(scaler.fit_transform(X))
```

```python
X1.head()
```

```python
fig, ax=plt.subplots(figsize=(18, 15))
ax=sns.heatmap(X1.corr(), annot=True)
plt.show()
```

```python
sns.heatmap(X.isna())
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
x_pca=pd.DataFrame(pca.fit_transform(X1))
x_pca.head()
```

```python
fig2, ax2=plt.subplots(figsize=(15, 13))
sns.heatmap(x_pca.corr(), ax=ax2, annot=True)
```

```python
exp_var=pca.explained_variance_ratio_
```

```python
print(exp_var)
```

```python
sum(exp_var[0:6])
```

```python
x_pca['target']=y
```

```python
x_pca.drop(columns=[i for i in range(7, 24)], inplace=True, axis=1)
```

```python
type(x_pca)
```

```python
x_pca=x_pca[x_pca['target'].notna()]
```

```python
x_pca.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y_=x_pca['target']
x_=x_pca.drop(columns=['target'], axis=1)
```

```python
fig1, ax1=plt.subplots(figsize=(10, 8))
sns.pairplot(x_)
```

```python
X_train, X_test, y_train, y_test=train_test_split(x_, y_, test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
pred=lin_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, pred, squared=False)
```
